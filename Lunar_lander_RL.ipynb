```python
import gymnasium as gym
from pyvirtualdisplay import Display

# Virtual display start karna
v_display = Display(visible=0, size=(1400, 900))
v_display.start()

print("✅ Virtual Display started!")
```

```python
from stable_baselines3 import PPO
from stable_baselines3.common.vec_env import DummyVecEnv
import gymnasium as gym

# 1. Lunar Lander game load karein (Version v3 use karein)
env_name = 'LunarLander-v3'
env = gym.make(env_name, render_mode="rgb_array")

# 2. Environment ko wrap karein
env = DummyVecEnv([lambda: env])

# 3. AI Model (PPO Algorithm)
model = PPO('MlpPolicy', env, verbose=1)

print("✅ Fixed! Version v3 loaded and model is ready.")
```

```python
 # 1. Setup the model with stable settings
model = PPO('MlpPolicy', env,
            verbose=1,
            learning_rate=0.0003,
            gamma=0.99)

# 2. Train for 500,000 steps to ensure it learns to land
print("🚀 Training starting... Look for 'ep_rew_mean' to become positive.")
model.learn(total_timesteps=500000)

# 3. Save the brain
model.save("ppo_lunar_lander_pro")
print("✅ Training Finished! Now you can run the evaluation code.")
```

```python
 from stable_baselines3.common.evaluation import evaluate_policy

# Test the model
mean_reward, std_reward = evaluate_policy(model, env, n_eval_episodes=10, deterministic=True)

print(f"📊 Average Score: {mean_reward}")
if mean_reward >= 200:
    print("🏆 Success! Your AI is now landing perfectly.")
else:
    print("ℹ️ If the score is still low, the AI needs even more steps.")
```

```python
import base64
from pathlib import Path
from IPython import display as ipythondisplay

def show_video(directory):
    html = []
    for mp4 in Path(directory).glob("*.mp4"):
        video_b64 = base64.b64encode(mp4.read_bytes()).decode()
        html.append(f'''<video alt="{mp4}" autoplay
                      loop controls style="height: 400px;">
                      <source src="data:video/mp4;base64,{video_b64}" type="video/mp4" />
                 </video>''')
    ipythondisplay.display(ipythondisplay.HTML(data="".join(html)))

print("✅ Video function ready!")
```

```python
from stable_baselines3.common.vec_env import VecVideoRecorder

# 1. Create a folder to save the video
video_folder = './videos/'
video_length = 500 # Number of steps to record

# 2. Setup the recorder
video_env = VecVideoRecorder(env, video_folder,
                             record_video_trigger=lambda x: x == 0,
                             video_length=video_length,
                             name_prefix="trained-agent-lunarlander")

# 3. Let the AI play
obs = video_env.reset()
for _ in range(video_length):
    action, _states = model.predict(obs, deterministic=True)
    obs, rewards, dones, info = video_env.step(action)

# 4. Save and Show the video
video_env.close()
show_video(video_folder)
```
