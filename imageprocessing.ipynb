```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("sample.jpg")
gray = cv2.imread("sample.jpg", 0)

if img is None or gray is None:
    raise FileNotFoundError("Place an image named sample.jpg in the working folder.")
```

```python
print("Color shape:", img.shape)
print("Gray shape:", gray.shape)
print("Gray pixel at (100,100):", gray[100, 100])
print("Color pixel at (100,100) in BGR:", img[100, 100])
```

```python
b, g, r = cv2.split(img)

fig, ax = plt.subplots(1, 4, figsize=(14, 4))
ax[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
ax[0].set_title("Original")
ax[1].imshow(b, cmap="gray")
ax[1].set_title("Blue Channel")
ax[2].imshow(g, cmap="gray")
ax[2].set_title("Green Channel")
ax[3].imshow(r, cmap="gray")
ax[3].set_title("Red Channel")
for a in ax:
    a.axis("off")
plt.show()
```

```python
bright = cv2.convertScaleAbs(img, alpha=1.0, beta=40)   # brightness
contrast = cv2.convertScaleAbs(img, alpha=1.5, beta=0)    # contrast

cv2.imwrite("bright.jpg", bright)
cv2.imwrite("contrast.jpg", contrast)
```

```python
negative = 255 - gray

_, thresh_binary = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
adaptive = cv2.adaptiveThreshold(gray, 255,
                                 cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                 cv2.THRESH_BINARY, 11, 2)
```

```python
equalized = cv2.equalizeHist(gray)

fig, ax = plt.subplots(1, 4, figsize=(16, 4))
ax[0].imshow(gray, cmap="gray")
ax[0].set_title("Original Gray")
ax[1].imshow(negative, cmap="gray")
ax[1].set_title("Negative")
ax[2].imshow(thresh_binary, cmap="gray")
ax[2].set_title("Binary Threshold")
ax[3].imshow(equalized, cmap="gray")
ax[3].set_title("Equalized")
for a in ax:
    a.axis("off")
plt.show()
```

```python
plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.hist(gray.ravel(), bins=256, range=[0, 256])
plt.title("Original Histogram")

plt.subplot(1, 2, 2)
plt.hist(equalized.ravel(), bins=256, range=[0, 256])
plt.title("Equalized Histogram")
plt.show()
```
