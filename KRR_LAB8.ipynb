```python
def infer_temperature_trend(heater_on: bool, window_open: bool) -> str:
    # 1. Base check: If heater is off, we assume it's stable/not increasing
    if not heater_on:
        return "stable"

    # 2. Check for the exception that overrides the default assumption
    if window_open:
        return "decreasing"  # Exception overrides default rule

    # 3. Default Reasoning rule
    return "increasing"  # Typical case rule


# --- Test Cases to Demonstrate Non-Monotonic Reasoning ---

# Test 1: Typical case (Default reasoning applies)
print (f"Heater ON, Window CLOSED -> Temperature is: {infer_temperature_trend(heater_on=True, window_open=False)}")
# Expected Output: increasing

# Test 2: Exception introduced (Previous conclusion is retracted/changed)
print(f"Heater ON, Window OPEN   -> Temperature is: {infer_temperature_trend(heater_on=True, window_open=True)}")
# Expected Output: decreasing
```
