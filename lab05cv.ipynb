```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load image in grayscale
img = cv2.imread("test.jpg", 0)

if img is None:
    print("Error: 'test.jpg' not found. Check filename/directory.")
else:
    print("Image loaded successfully.")
```

```python
# Create a Gaussian Pyramid
layer = img.copy()
gaussian_pyr = [layer]

# Generate 3 levels of the pyramid
for i in range(3):
    layer = cv2.pyrDown(layer)
    gaussian_pyr.append(layer)

# Display the pyramid levels
plt.figure(figsize=(15, 5))
for i in range(len(gaussian_pyr)):
    plt.subplot(1, 4, i+1)
    plt.imshow(gaussian_pyr[i], cmap='gray')
    plt.title(f"Level {i}")
    plt.axis('off')
plt.show()
```
