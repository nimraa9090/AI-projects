```python
# 1. THE FACT BASE (What we know right now)
fact_base = {
    "studied": True,
    "attended_class": True,
    "paid_fees": True,

    # These start as False because the AI hasn't proven them yet
    "passed_exam": False,
    "earned_credits": False,
    "can_graduate": False
}

print("--- Starting AI Engine ---")
print(f"Initial Facts: Studied={fact_base['studied']}, Attended={fact_base['attended_class']}, Paid Fees={fact_base['paid_fees']}\n")


# 2. THE INFERENCE ENGINE (Running the rules forward)
# We loop a few times to let conclusions chain together
for loop in range(3):
    print(f"Scanning Rules (Pass {loop + 1})...")

    # RULE 1: IF you studied AND attended class -> THEN you passed the exam
    if fact_base["studied"] and fact_base["attended_class"] and not fact_base["passed_exam"]:
        fact_base["passed_exam"] = True
        print("➔ Rule 1 Fired! Concluded: You passed the exam.")

    # RULE 2: IF you passed the exam -> THEN you earned your credits
    if fact_base["passed_exam"] and not fact_base["earned_credits"]:
        fact_base["earned_credits"] = True
        print("➔ Rule 2 Fired! Concluded: You earned college credits.")

    # RULE 3: IF you earned credits AND paid your fees -> THEN you can graduate
    if fact_base["earned_credits"] and fact_base["paid_fees"] and not fact_base["can_graduate"]:
        fact_base["can_graduate"] = True
        print("➔ Rule 3 Fired! Concluded: YOU CAN GRADUATE!")

print("\n--- Final Results ---")
print(f"Can the student graduate? {fact_base['can_graduate']}")
```
