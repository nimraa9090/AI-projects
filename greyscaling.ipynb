```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("sample.jpg")
if img is None:
    raise FileNotFoundError("Place an image named sample.jpg in the working folder.") [cite: 40]
```

```python
print("Shape:", img.shape) # (height, width, channels)
print("Data type:", img.dtype)
print("Top-left pixel (BGR):", img[0, 0])
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(6, 4))
plt.imshow(img_rgb)
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
resized = cv2.resize(img, (500, 300))
cropped = img[50:250, 100:350]

cv2.imwrite("resized.jpg", resized)
cv2.imwrite("cropped.jpg", cropped)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imwrite("gray.jpg", gray)

plt.figure(figsize=(5, 4))
plt.imshow(gray, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")
plt.show()
```

```python
flipped = cv2.flip(img, 1)  # horizontal flip

(h, w) = img.shape[:2]
center = (w // 2, h // 2)
M = cv2.getRotationMatrix2D(center, 30, 1.0)
rotated = cv2.warpAffine(img, M, (w, h))

cv2.imwrite("flipped.jpg", flipped)
cv2.imwrite("rotated.jpg", rotated)
```

```python
annotated = img.copy()
cv2.rectangle(annotated, (40, 40), (220, 160), (0, 255, 0), 2)
cv2.putText(annotated, "Sample ROI", (45, 35),
            cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 0, 0), 2)

cv2.imwrite("annotated.jpg", annotated)

plt.figure(figsize=(6, 4))
plt.imshow(cv2.cvtColor(annotated, cv2.COLOR_BGR2RGB))
plt.title("Annotated Output")
plt.axis("off")
plt.show()
```
