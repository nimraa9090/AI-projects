```python
# THIS IS THE FIRST CELL
article_text = """
Artificial intelligence is transforming the way we live and work in the modern world.
Machine learning algorithms are being used to analyze vast amounts of data, helping
scientists discover new medicines and engineers build more efficient infrastructure.
In the field of education, AI-powered tools can personalize learning experiences,
allowing students to progress at their own pace. Meanwhile, in the corporate sector,
automation is streamlining routine tasks, which enables employees to focus on
creative and strategic problem-solving. Despite these benefits, the rise of
advanced technology also raises important ethical questions regarding privacy,
data security, and the future of the job market. It is crucial that we develop
clear guidelines and regulatory frameworks to ensure these technologies are used
responsibly. As we move forward, balancing innovation with human-centric values
will be the key to successfully integrating AI into our daily lives. By prioritizing
transparency and collaboration, we can harness the power of artificial intelligence
to build a more sustainable and equitable future for everyone.
"""
# Paste your long paragraph here inside these triple quotes.
# The code will use this variable to perform the summarization.
```

```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
```

```python
import re
import heapq
import nltk

# NLTK sentence tokenizer is used for splitting text into sentences
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
```

```python
# Assuming 'article_text' is your source string

# Clean the text
article_text = re.sub(r'\[[0-9]*\]', ' ', article_text)
article_text = re.sub(r'\s+', ' ', article_text)
# Remove special characters and digits
formatted_article_text = re.sub('[^a-zA-Z]', ' ', article_text)
formatted_article_text = re.sub(r'\s+', ' ', formatted_article_text)
```

```python
import nltk
nltk.download('punkt_tab')
sentence_list = sent_tokenize(article_text)
```

```python
stopwords = nltk.corpus.stopwords.words('english')

word_frequencies = {}
for word in word_tokenize(formatted_article_text):
    if word not in stopwords:
        if word not in word_frequencies.keys():
            word_frequencies[word] = 1
        else:
            word_frequencies[word] += 1

# Normalize frequencies
maximum_frequncy = max(word_frequencies.values())
for word in word_frequencies.keys():
    word_frequencies[word] = (word_frequencies[word]/maximum_frequncy)
```

```python
sentence_scores = {}
for sent in sentence_list:
    for word in word_tokenize(sent.lower()):
        if word in word_frequencies.keys():
            if len(sent.split(' ')) < 30: # Only consider sentences under 30 words
                if sent not in sentence_scores.keys():
                    sentence_scores[sent] = word_frequencies[word]
                else:
                    sentence_scores[sent] += word_frequencies[word]
```

```python
summary_sentences = heapq.nlargest(5, sentence_scores, key=sentence_scores.get)
summary = ' '.join(summary_sentences)
print(summary)
```
