```python
import pandas as pd
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
```

```python
data = [10, 12, 14, 15, 16, 18, 20, 22, 100]

df = pd.DataFrame(data, columns=['Values'])

print("Dataset:")
print(df)
```

```python
z_scores = np.abs(stats.zscore(df['Values']))

print("Z-Scores:")
print(z_scores)
```

```python
outliers = df[z_scores > 2]

print("Detected Outliers:")
print(outliers)
```

```python
plt.figure(figsize=(6,4))
plt.boxplot(df['Values'])
plt.title('Outlier Detection using Z-Score')
plt.ylabel('Values')
plt.show()
```

```python
print("Observation:")
print("The value 100 is far away from the other values.")
print("Therefore, 100 is detected as an outlier.")
```
