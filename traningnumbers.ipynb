```python
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plt

# Step 1: Load and preprocess the MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0

# Step 2: Define the model architecture
model = models.Sequential([
    layers.Flatten(input_shape=(28, 28)),           # Flatten the 28x28 images into 1D arrays
    layers.Dense(128, activation='relu'),           # Dense layer with 128 neurons and ReLU activation
    layers.Dropout(0.2),                             # Dropout layer to prevent overfitting
    layers.Dense(10, activation='softmax')          # Output layer with 10 neurons for each digit (0-9)
])

# Step 3: Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Step 4: Train the model
model.fit(x_train, y_train, epochs=5)

# Step 5: Evaluate the model
test_loss, test_acc = model.evaluate(x_test, y_test)
print('Test accuracy:', test_acc)

# Step 6: Make predictions
predictions = model.predict(x_test)

# Step 7: Display the first 25 test images and their predicted labels
plt.figure(figsize=(10, 10))
j = 224
for i in range(25):
    plt.subplot(5, 5, i + 1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(x_test[j], cmap=plt.cm.binary)
    predicted_label = predictions[j].argmax()
    true_label = y_test[j]
    color = 'green' if predicted_label == true_label else 'red'
    plt.xlabel(f'Predicted: {predicted_label}, True: {true_label}', color=color)
    j += 1
plt.tight_layout()
plt.show()
```

```python
def add_student(name, score):
    with open("student_data.txt", "a") as f:
        f.write(f"{name},{score}\n")
def display_students():
    with open("student_data.txt", "r") as f:
        print(f.read())
add_student("Ali", 88)
add_student("Sara", 92)
display_students()
```

```python
# Create the text content
text1_content = """Amanullah Baloch s/o Ghulam Akbar Baloch,
HAZARA University,
Khyber Pakhtunkhawa
District Mansehra."""

text2_content = """Ahsanullah Baloch s/o Ghulam Akbar Baloch,
GOMAL University,
Khyber Pakhtunkhawa
District Dera Ismail Khan."""

# Simulate writing to text files (you can skip this if files already exist)
with open("Text_1.txt", "w") as f1:
    f1.write(text1_content)

with open("Text_2.txt", "w") as f2:
    f2.write(text2_content)

# Read and process the files
def clean_text(filepath):
    with open(filepath, "r") as file:
        content = file.read()
    # Remove punctuation and convert to lowercase
    for ch in [',', '.', '\n']:
        content = content.replace(ch, ' ')
    return set(content.lower().split())

# Extract word sets
set1 = clean_text("Text_1.txt")
set2 = clean_text("Text_2.txt")

# Find intersection
common_words = set1 & set2

print("Common words:", common_words)
```

```python
# Ask the user how many elements they want to enter
n = int(input("How many numbers do you want to enter? "))

# Input the numbers one by one
numbers = []
for i in range(n):
    num = float(input(f"Enter number {i + 1}: "))
    numbers.append(num)

# Calculate sum and average
total = sum(numbers)
average = total / len(numbers)

# Display the result
print("Sum of numbers:", total)
print("Average of numbers:", average)
```
